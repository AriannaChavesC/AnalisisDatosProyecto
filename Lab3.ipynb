```python
#Preguntas Netflix :
netflix_data = df[df['streaming_service'] == 'netflix']

#Pregunta1:
# Agrupar por la columna 'main_genre' y contar el n√∫mero de pel√≠culas en cada grupo
genre_counts = netflix_data.groupby('main_genre').size().reset_index(name='number_of_movies')
# Esta consulta agrupa los generos principales por su cantidad de peliculas en la plataforma
genre_counts

```

```python
#Pregunta 2:
popularity_by_genre = netflix_data.groupby('main_genre')['tmdb_popularity'].mean().reset_index()
# Ordenar por popularidad en orden descendente
popularity_by_genre_sorted = popularity_by_genre.sort_values(by='tmdb_popularity', ascending=False).reset_index(drop=True)
# Esta consulta agrupa por genero principal y se orden de forma descendente por popularidad en la plataforma (Top 10 Generos mas populares en netflix)
popularity_by_genre_sorted

```

```python
#Preguntas darkmatter
darkmatter_data = df[df['streaming_service'] == 'darkmatter']

#Pregunta 3:
score_by_genre = darkmatter_data.groupby('main_genre')['tmdb_score'].mean().reset_index()
# Ordenar por calificacion en orden descendente
score_by_genre_sorted = score_by_genre.sort_values(by='tmdb_score', ascending=False).reset_index(drop=True)
# Esta consulta muestra de forma ascendente los generos por su calificacion (Top generos mejores calificados en darkmatter tv)
score_by_genre_sorted

```

```python
#Pregunta 4:
top_10_scores = darkmatter_data[['title', 'tmdb_score', 'tmdb_popularity']]
# Ordenar por tmdb_score en orden descendente y tomar los primeros 10 resultados y se muestran an relacion a su popularidad
top_10_scores_sorted = top_10_scores.sort_values(by='tmdb_score', ascending=False).head(10)
# Esta consulta provee el top de peliculas mejores calificadas en relacion a su popularidad
top_10_scores_sorted
```

```python

# Filtrar los datos para Amazon
amazon_df = df[df['streaming_service'] == 'amazon']

# Agrupar por g√©nero y calcular el promedio del rating en Amazon
rating_amazon = amazon_df.groupby('main_genre')['tmdb_score'].mean()

# Ordenar los g√©neros por rating promedio en orden descendente y obtener el top 10
top10_amazon = rating_amazon.sort_values(ascending=False).head(10)

print("Top 10 g√©neros con el rating m√°s alto en Amazon:")
top10_amazon
```

```python

# Agrupar por a√±o de lanzamiento y encontrar el t√≠tulo con el rating m√°s alto
mejor_pelicula_por_ano_amazon = amazon_df.loc[amazon_df.groupby('release_year')['tmdb_score'].idxmax()]

# Seleccionar las columnas relevantes y agregar el g√©nero
mejor_pelicula_por_ano_amazon = mejor_pelicula_por_ano_amazon[['release_year', 'title', 'tmdb_score', 'genres']]

# Ordenar por el rating en orden descendente y obtener el top 10
top10_anos_amazon = mejor_pelicula_por_ano_amazon.sort_values(by='tmdb_score', ascending=False).head(10)

print("Top 10 a√±os con las pel√≠culas con el mayor rating en Amazon:")
top10_anos_amazon

```

```python

# Filtrar los datos para Hulu
hulu_df = df[df['streaming_service'] == 'hbo']  # Ajusta 'hbo' si es necesario

# Agrupar por g√©nero y calcular el promedio del rating en Hulu
rating_hulu = hulu_df.groupby('main_genre')['tmdb_score'].mean()

# Ordenar los g√©neros por rating promedio en orden ascendente y obtener el top 10
top10_hulu = rating_hulu.sort_values(ascending=True).head(10)

print("Top 10 g√©neros con el rating m√°s bajo en Hulu:")
top10_hulu

```

```python

# Asegurarse de que 'genres' es una lista o cadena que puede ser procesada
# Si 'genres' es una cadena, convierte las cadenas de g√©neros a listas
if isinstance(hulu_df['genres'].iloc[0], str):
    hulu_df['genres'] = hulu_df['genres'].apply(lambda x: eval(x) if isinstance(x, str) else x)

# Explode the genres column to have one genre per row
hulu_df_exploded = hulu_df.explode('genres')

# Contar el n√∫mero de pel√≠culas por g√©nero
numero_peliculas_por_genero = hulu_df_exploded['genres'].value_counts()

print("N√∫mero de pel√≠culas por g√©nero en Hulu:")
numero_peliculas_por_genero
```
